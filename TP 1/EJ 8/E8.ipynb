```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Función para ajustar la ley de Heaps
def heaps_law(n, k, beta):
    return k * (n ** beta)
```

```python
# Cargar los datos del archivo de salida generado por el script anterior
with open('output.txt', 'r') as file:
    lines = file.readlines()
    total_terms_processed = int(lines[0].split(':')[1])
    unique_terms = int(lines[1].split(':')[1])

print("Total de términos procesados:", total_terms_processed)
print("Términos únicos:", unique_terms)
```

```python
# Ajustar a la ley de Heaps
x = np.array([i for i in range(1, total_terms_processed + 1)])
y = np.array([unique_terms])
popt, pcov = curve_fit(heaps_law, x, y[0])
```

```python
# Graficar los datos y el ajuste de la ley de Heaps
y = [unique_terms] * total_terms_processed
plt.scatter(x, y, label='Datos reales')
plt.plot(x, heaps_law(x, *popt), color='red', label='Ajuste de Heaps')
plt.xlabel('Términos totales procesados')
plt.ylabel('Términos únicos')
plt.title('Ley de Heaps')
plt.legend()
plt.show()
```

```python
# Graficar los ajustes variando los parámetros de la expresión
plt.figure(figsize=(12, 6))
for i, beta in enumerate(np.linspace(0.5, 1.5, 5)):
    plt.subplot(2, 3, i+1)
    plt.scatter(x, y, label='Datos reales')
    plt.plot(x, heaps_law(x, popt[0], beta), color='red', label=f'Ajuste (beta={beta:.2f})')
    plt.xlabel('Términos totales procesados')
    plt.ylabel('Términos únicos')
    plt.title(f'Ley de Heaps (Ajuste con beta={beta:.2f})')
    plt.legend()

plt.tight_layout()
plt.show()
```
