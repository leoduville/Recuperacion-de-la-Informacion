```python
import pyterrier as pt
from pathlib import Path
if not pt.started():
    pt.init()
```

```python
files = pt.io.find_files("en")
files
```

```python
indexer = pt.FilesIndexer("C:/Users/leo_2/OneDrive/Documentos/GitHub/Recuperacion de la Informacion/Recuperacion-de-la-Informacion/TP 2/EJ 5/custom_trec_index", verbose=True, overwrite=True, meta={"docno":20, "filename":512})
indexref = indexer.index(files)
```

```python
index = pt.IndexFactory.of(indexref)
print(index.getCollectionStatistics().toString())
```

```python
# QUERY 1
tf_idf = pt.BatchRetrieve(index, num_results=10, wmodel="TF_IDF", metadata=["filename"])
tf_idf_results = tf_idf.search("software")
tf_idf_results
```

```python
bm25 = pt.BatchRetrieve(index, num_results=10, wmodel="BM25", metadata=["filename"])
bm25_results = bm25.search("software")
bm25_results
```

```python
# Extraer los docno de los documentos y sus posiciones en los resultados de TF-IDF
tf_idf_results = tf_idf_results.loc[tf_idf_results['query'] == 'software']  # Filtrar resultados para la consulta 'software'
tf_idf_results = tf_idf_results.sort_values(by='rank')  # Ordenar por la posición en los resultados
tf_idf_docnos = tf_idf_results['filename'].tolist()  # Lista de docno en orden de posición
print(tf_idf_docnos)
```

```python
# Hacer lo mismo para BM25
bm25_results = bm25_results.loc[bm25_results['query'] == 'software']  # Filtrar resultados para la consulta 'software'
bm25_results = bm25_results.sort_values(by='rank')  # Ordenar por la posición en los resultados
bm25_docnos = bm25_results['filename'].tolist()  # Lista de docno en orden de posición
print(bm25_docnos)
```

```python
from scipy.stats import spearmanr

# Calcular el coeficiente de correlación de Spearman
spearman_coefficient, _ = spearmanr(tf_idf_docnos, bm25_docnos)

print("Coeficiente de correlación de Spearman:", spearman_coefficient)
```
